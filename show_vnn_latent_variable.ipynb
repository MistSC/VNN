```python
lly=np.load('mnist_vnn_label.npy')
ldcz=np.load('mnist_vnn_dcz.npy')
lecz=np.load('mnist_vnn_ecz.npy')
```

```python
X = lecz
labels = lly
Y = tsne(X, 2, 5, 30.0);
```

```python
np.save('tsne_encode_z.npy',Y)
```

```python
Plot.figure(1)
for i in np.where(lly==0):
    Plot.scatter(Y[i,0], Y[i,1], 0.1, lly[i], label=0, color='r')
for i in np.where(lly==1):
    Plot.scatter(Y[i,0], Y[i,1], 0.1, lly[i], label=1, color='b')
for i in np.where(lly==2):
    Plot.scatter(Y[i,0], Y[i,1], 0.1, lly[i], label=2, color='g')
for i in np.where(lly==3):
    Plot.scatter(Y[i,0], Y[i,1], 0.1, lly[i], label=3, color='c')
for i in np.where(lly==4):
    Plot.scatter(Y[i,0], Y[i,1], 0.1, lly[i], label=4, color='m')
for i in np.where(lly==5):
    Plot.scatter(Y[i,0], Y[i,1], 0.1, lly[i], label=5, color='k')
for i in np.where(lly==6):
    Plot.scatter(Y[i,0], Y[i,1], 0.1, lly[i], label=6, color='y')
for i in np.where(lly==7):
    Plot.scatter(Y[i,0], Y[i,1], 0.1, lly[i], label=7, color=(0.5,0.5,0.5))
for i in np.where(lly==8):
    Plot.scatter(Y[i,0], Y[i,1], 0.1, lly[i], label=8, color=(0.3,0.3,0.3))
for i in np.where(lly==9):
    Plot.scatter(Y[i,0], Y[i,1], 0.1, lly[i], label=9, color=(1,0.8,0.8))
Plot.title('t-SNE of encode latent variable')
Plot.legend(bbox_to_anchor=(1.2,0.9))
Plot.grid(True)
Plot.show()
```
